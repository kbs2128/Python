```python
def feeder(f):
    x=test_case["Meter Name"].where(test_case["Feeder Way CoR Number"]==f)
    x=x.dropna()
    x=list(x)
    x.insert(0,"Time Stamp")
    df=all_meter_data.loc[:,x]
    df["Sum"]=0
    for j in x[1:]:
        df["Sum"]+=df[j]
    print(df)

```

```python
def sub_station(s):
    f=test_case["Feeder Way CoR Number"].where(test_case["Secondary Substation CoR Number"]==s)
    f=f.dropna()
    f=list(f)
    f=set(f)

    for i in f:
        x=test_case["Meter Name"].where(test_case["Feeder Way CoR Number"]==i)
        x=x.dropna()
        x=list(x)
        x.insert(0,"Time Stamp")
        df=all_meter_data.loc[:,x]
        df["Sum"]=0
        for j in x[1:]:
            df["Sum"]+=df[j]
        print("\n")
        print(i+ " Feeder")
        print("\n")
        print(df)
```

```python
sub_station("WOOLWORTHS- IPSWICH")
```

```python
sub_station("FORE HAMLET LOCAL RMU's")
```

```python
sub_station("DERBY ROAD")
```

```python
sub_station("FELIXSTOWE RD")
```
